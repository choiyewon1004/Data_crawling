```python
import pandas as pd
import operator
import matplotlib.pylab as plt
```

```python
def count_frequency(my_list):
    
    count = {}
    
    for item in my_list:
        count[item] = count.get(item, 0) + 1
        
    return count
```

```python
def order_dic(my_dic):
    return sorted(
        my_dic.items(),
        key= operator.itemgetter(1),
        reverse = True
    )
    
```

```python
for mon in range(1,13):
    df = pd.read_csv(f"month_final/{mon}_item_list.csv")
    t_li = list(df['item_type'])
    count = count_frequency(t_li)
    count = order_dic(count)
    
    print(f"!!!!! {mon} count !!!!")
    print(count)
    print()
    print()
    print()
```
